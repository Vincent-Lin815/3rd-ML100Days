```python
import os
import keras
import itertools
from keras.layers import Dropout
import matplotlib.pyplot as plt
%matplotlib inline
# Disable GPU
os.environ["CUDA_VISIBLE_DEVICES"] = ""
```

```python
train, test = keras.datasets.cifar10.load_data()
```

```python
## 資料前處理
def preproc_x(x, flatten=True):
    x = x / 255.
    if flatten:
        x = x.reshape((len(x), -1))
    return x

def preproc_y(y, num_classes=10):
    if y.shape[-1] == 1:
        y = keras.utils.to_categorical(y, num_classes)
    return y    
```

```python
x_train, y_train = train
x_test, y_test = test

# Preproc the inputs
x_train = preproc_x(x_train)
x_test = preproc_x(x_test)

# Preprc the outputs
y_train = preproc_y(y_train)
y_test = preproc_y(y_test)
```

```python

def build_mlp(input_shape, output_units = 10, num_neurons = [512, 256, 128], drop_ratio = 0.3):
    """Code Here
    建立神經網路
    """
    input_layer = keras.layers.Input(input_shape)
    
    for i, n_units in enumerate(num_neurons):
        if i == 0:
            x = keras.layers.Dense(units = n_units,
                                   activation = 'relu', 
                                   name = 'hidden_layer'+str(i+1))(input_layer)
        else:
            x = keras.layers.Dense(units = n_units,
                                   activation = 'relu', 
                                   name = 'hidden_layer'+str(i+1))(x)
            x = Dropout(drop_ratio)(x)
    output_layer = keras.layers.Dense(units = output_units, activation = 'softmax', name = 'output_layer')(x)
    model = keras.models.Model(inputs = [input_layer], outputs = [output_layer])
    return model
```

```python
"""Code Here
設定超參數
"""
LEARNING_RATE = 1e-3
EPOCHS = 10
BATCH_SIZE = 256
MOMEMTUM = 0.95
DROPOUT_EXP = [0.25, 0.4, 0.55, 0.7, 0.85]
```

```python
results = {}
"""Code Here
撰寫訓練流程並將結果用 dictionary 紀錄
"""
for drop_ratio in DROPOUT_EXP:
    keras.backend.clear_session()
    print('Experiment with drop_ratio:{:.2f}'.format(drop_ratio))
    model = build_mlp(input_shape = x_train.shape[1:], output_units = 10, num_neurons = [512, 256, 128], drop_ratio = drop_ratio)
    model.summary()
    optimizer = keras.optimizers.Adam(lr=LEARNING_RATE)
    model.compile(loss = 'categorical_crossentropy', metrics = ['accuracy'], optimizer = optimizer)
    
    model.fit(x_train, y_train, 
              epochs = EPOCHS, 
              batch_size = BATCH_SIZE, 
              validation_data = (x_test, y_test),
              shuffle = True)
    
    # Collect results
    train_loss = model.history.history["loss"]
    valid_loss = model.history.history["val_loss"]
    train_acc = model.history.history["accuracy"]
    valid_acc = model.history.history["val_accuracy"]
    
    exp_name_tag = "exp-l1-%s" % str(drop_ratio)
    results[exp_name_tag] = {'train-loss': train_loss,
                             'valid-loss': valid_loss,
                             'train-acc': train_acc,
                             'valid-acc': valid_acc}
    
    plt.plot(range(len(train_loss)), train_loss, label="train loss")
    plt.plot(range(len(valid_loss)), valid_loss, label="valid loss")
    plt.legend()
    plt.title("Loss")
    plt.show()
    
    plt.plot(range(len(train_acc)), train_acc, label="train accuracy")
    plt.plot(range(len(valid_acc)), valid_acc, label="valid accuracy")
    plt.legend()
    plt.title("Accuracy")
    plt.show()
```
